```python
# =====================================================
# EXPERIMENT 1 : Python Environment Setup and Dataset Demo
# =====================================================

# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
from ipywidgets import interact

print("✅ Python environment ready!")


# =====================================================
# Step 2: Create and Load Dataset (No upload needed)
# =====================================================
csv_data = """Name,Age,Marks
Asha,22,85
Bala,23,90
Chetan,21,78
Divya,24,88
Eshan,22,95
Farah,23,82
Gopal,22,80
Heena,21,91
Irfan,24,87
Jaya,23,89
"""

# Load CSV data from string into a DataFrame
df = pd.read_csv(StringIO(csv_data))
print("\n✅ Dataset created and imported successfully!\n")
print(df.head())


# =====================================================
# Step 3: Display Basic Dataset Information
# =====================================================
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- Summary Statistics ---")
print(df.describe())


# =====================================================
# Step 4: Perform a Simple Operation + Condition Check
# =====================================================
# Find mean of first numeric column and filter rows above mean
num_col = df.select_dtypes(include='number').columns[0]
mean_value = df[num_col].mean()

print(f"\nMean of '{num_col}': {mean_value:.2f}")

# Condition check: values greater than mean
filtered_df = df[df[num_col] > mean_value]
print(f"\n✅ Rows where {num_col} > mean:\n")
print(filtered_df)


# =====================================================
# Step 5: Visualization using Matplotlib
# =====================================================
plt.figure(figsize=(6,4))
plt.bar(df['Name'], df[num_col], color='skyblue', edgecolor='black')
plt.title(f"Bar Chart of {num_col}")
plt.xlabel("Name")
plt.ylabel(num_col)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# =====================================================
# Step 6:
```

```python
# =====================================================
# EXPERIMENT 2 : Data Import and Export
# =====================================================

# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO

print("✅ Libraries imported successfully!")


# =====================================================
# Step 2: Create a Sample Dataset (No external file needed)
# =====================================================
csv_data = """Name,Age,Marks
Asha,22,85
Bala,23,90
Chetan,21,78
Divya,24,88
Eshan,22,95
Farah,23,82
Gopal,22,80
Heena,21,91
Irfan,24,87
Jaya,23,89
"""

df = pd.read_csv(StringIO(csv_data))
print("\n✅ Original dataset created:\n")
print(df)


# =====================================================
# Step 3: Save Dataset to a CSV File (Export)
# =====================================================
df.to_csv("student_data.csv", index=False)
print("\n💾 Dataset exported to 'student_data.csv' successfully!\n")


# =====================================================
# Step 4: Import Dataset Back (Read from File)
# =====================================================
df_imported = pd.read_csv("student_data.csv")
print("✅ Dataset imported again successfully!\n")
print(df_imported.head())


# =====================================================
# Step 5: Perform Basic Data Operations
# =====================================================
# Add a new column "Result" based on condition
df_imported['Result'] = np.where(df_imported['Marks'] >= 85, 'Pass', 'Fail')

# Calculate Average Marks
average_marks = df_imported['Marks'].mean()
print(f"\n📊 Average Marks: {average_marks:.2f}")

print("\n✅ Updated Dataset with Result column:\n")
print(df_imported)


# =====================================================
# Step 6: Export Processed Data
# =====================================================
df_imported.to_csv("processed_student_data.csv", index=False)
print("\n💾 Processed data exported to 'processed_student_data.csv' successfully!")


# =====================================================
# Step 7: Simple Visualization
# =====================================================
plt.figure(figsize=(6,4))
plt.bar(df_imported['Name'], df_imported['Marks'], color='lightgreen', edgecolor='black')
plt.title("Students Marks Chart")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# =====================================================
# Step 8: Verification of Exported File
# =====================================================
print("\n✅ File 'processed_student_data.csv' contains the following data:\n")
print(pd.read_csv("processed_student_data.csv"))
```
