```python
import sqlite3
import pandas as pd

# Create in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create a table and insert some sample rows
cursor.execute('''
CREATE TABLE products (
    id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
)
''')

cursor.executemany('INSERT INTO products (name, price) VALUES (?, ?)', [
    ('Pen', 10.0),
    ('Notebook', 25.5),
    ('Eraser', 5.75)
])
conn.commit()

# Read from the SQL table
df_sql = pd.read_sql('SELECT * FROM products', conn)

print("SQL Data (Products Table):\n", df_sql)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('/content/sample_data/StudentsPerformance - StudentsPerformance.csv')

numeric_col=df.select_dtypes(include='number').columns.tolist()

plt.figure(figsize=(6 * len(numeric_col), 4))
for i, col in enumerate(numeric_col):
    plt.subplot(1, len(numeric_col), i + 1)
    sns.histplot(df[col], kde=True, color='blue')
    plt.title(f'Original {col}')
plt.tight_layout()
plt.show()

scaler=MinMaxScaler()
df_minmax=df.copy()
for cols in numeric_col:
  df_minmax[cols]=scaler.fit_transform(df_minmax[[cols]])

plt.figure(figsize=(6 * len(numeric_col), 4))
for i, col in enumerate(numeric_col):
    plt.subplot(1, len(numeric_col), i + 1)
    sns.histplot(df_minmax[col], kde=True, color='orange')
    plt.title(f'Original {col}')
plt.tight_layout()
plt.show()

df_standard = df.copy()
scaler = StandardScaler()
for col in numeric_col:
    df_standard[col] = scaler.fit_transform(df_standard[[col]])

plt.figure(figsize=(6 * len(numeric_col), 4))
for i, col in enumerate(numeric_col):
    plt.subplot(1, len(numeric_col), i + 1)
    sns.histplot(df_standard[col], kde=True, color='pink')
    plt.title(f'Original {col}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('/content/sample_data/StudentsPerformance - StudentsPerformance.csv')
df.head(5)
df.tail(5)

filtered = df[df["math score"] > 90]
filtered.shape

filtered=df[(df["math score"] > 90) & (df["reading score"] > 90)]
filtered.shape

subset=df[["math score","reading score","writing score"]]
subset.head()
```

```python
##Descriptive Statistics
mode_val=df['math score'].mode()[0]
mode_val

mean_val=df['math score'].mean()
mean_val

median_val=df['math score'].median()
median_val

variance=df['math score'].var()
variance

std=df['math score'].std()
std

import matplotlib.pyplot as plt
plt.hist(df["math score"], bins=10, color="skyblue", edgecolor="black")

# Add vertical lines
plt.axvline(mean_val, color="red", linestyle="-", linewidth=2, label=f"Mean: {mean_val:.2f}")
plt.axvline(median_val, color="green", linestyle="--", linewidth=2, label=f"Median: {median_val:.2f}")
plt.axvline(mode_val, color="blue", linestyle="--", linewidth=2, label=f"Mode: {mode_val}")

# Labels & Title
plt.title("Math Score Distribution with Mean, Median, and Mode")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.legend()

# Show plot
plt.show()
```

```python
#Exp 5
import pandas as pd
df = pd.read_csv("/content/sample_data/StudentsPerformance - StudentsPerformance.csv", encoding='latin-1')
df.head()

import matplotlib.pyplot as plt
df_head = df.head(10)
plt.plot(df_head.index, df_head["math score"], marker='o', label="Math")
plt.plot(df_head.index, df_head["reading score"], marker='s', label="Reading")
plt.plot(df_head.index, df_head["writing score"], marker='^', label="Writing")
plt.title("Line Chart of Scores (First 10 Students)")
plt.xlabel("Student Index")
plt.ylabel("Scores")
plt.legend()
plt.show()

avg_scores = df.groupby("gender")[["math score","reading score","writing score"]].mean()
avg_scores.plot(kind="bar", figsize=(6,4))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.show()

plt.hist(df["math score"], bins=15, color="green", edgecolor="black")
plt.title("Histogram of Math Scores")
plt.xlabel("Math Score Range")
plt.ylabel("Frequency")
plt.show()
```
